### Module Imports

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix

from sklearn.tree import DecisionTreeClassifier

from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image 
```

## Iris Data

*Here I am not dividing dataset into testing, and training data, as our goal is to build* ***Decision Tree***

```python
iris = datasets.load_iris()

X = iris.data
Y = iris.target
```

### Decision Tree Classifier

```python
clf = DecisionTreeClassifier()
clf.fit(X, Y)
```

# Visualising Decision Tree

## Create Dot Data Object from Decision Tree Classifier

```python
# dot data object
dot_data_obj = export_graphviz(clf, out_file = None,
                           feature_names=iris.feature_names,
                           class_names=iris.target_names,
                           filled=True, rounded=True)

dot_data_obj
```

## Create Graph Object from Dot Data Object

```python
# graph object
graph_obj = pydotplus.graph_from_dot_data(dot_data_obj)

graph_obj
```

## Saving Tree in pdf

```python
# saving graph object in pdf
graph_obj.write_pdf("iris_decision_tree.pdf")
```

## Display Image of Decision Tree

```python
# display image of graph
Image(graph_obj.create_png())
```

### Predictions

```python
Y_pred = clf.predict(X)
```

### Confusion Matrix

```python
confusion_matrix(Y, Y_pred)
```
