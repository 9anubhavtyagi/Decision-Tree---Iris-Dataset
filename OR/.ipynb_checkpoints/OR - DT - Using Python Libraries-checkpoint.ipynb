### Module Imports

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

from sklearn.metrics import confusion_matrix

from sklearn.tree import DecisionTreeClassifier

from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Image

import numpy as np
```

### XOR Data

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 1, 1, 1])
```

### Decision Tree Classifier

```python
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X, Y)
```

# Visualising Decision Tree

## Create Dot Data Object from Decision Tree Classifier

```python
# dot data object
dot_data_obj = export_graphviz(dt_clf, out_file = None, filled=True, rounded=True,
                               feature_names = ['X1', 'X2'],
                               class_names = ['False', 'True'])
dot_data_obj
```

## Create Graph Object from Dot Data Object

```python
# graph object
graph_obj = pydotplus.graph_from_dot_data(dot_data_obj)

graph_obj
```

## Saving Tree in pdf

```python
# saving graph object in pdf
graph_obj.write_pdf("or_decision_tree.pdf")
```

## Display Image of Decision Tree

```python
# display image of graph
Image(graph_obj.create_png())
```

### Confusion Matrix

```python
Y_pred = dt_clf.predict(X)

confusion_matrix(Y, Y_pred)
```
